```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Read the Excel file containing the list of websites
df = pd.read_excel('input.xlsx')

# Iterate over each URL in the dataframe
for idx, row in df.iterrows():
    url = row['URL']
    article_id = row['URL_ID']
    
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        html_content = response.content
        soup = BeautifulSoup(html_content, 'html.parser')
        
        # Find the website title
        website_title = soup.title.string
        
        # Find the paragraph content inside the specified div
        div_with_paragraph = soup.find('div', class_='td-post-content tagdiv-type')
        if div_with_paragraph:
            paragraph_content = div_with_paragraph.find('p').get_text(strip=True)
        else:
            paragraph_content = "Paragraph content not found."
        
        # Print or save the title and paragraph content
        print(f"Website Title ({article_id}):", website_title)
        print(f"Paragraph Content ({article_id}):", paragraph_content)
        
        # Save the extracted text into a text file
        """with open(f'{article_id}.txt', 'w') as f:
            f.write(website_title + '\n')
            f.write(paragraph_content)"""
    else:
        print(f"Failed to retrieve data from URL ({url}). Status code: {response.status_code}")
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from textblob import TextBlob
import string
import os

# Load stop words
stop_words_path = 'StopWords'
stop_words_files = os.listdir(stop_words_path)
stop_words = set()

# Iterate over each file in the stop_words directory
for file in stop_words_files:
    if file.endswith('.txt'):
        with open(os.path.join(stop_words_path, file), 'r', encoding='latin-1') as f:
            stop_words.update(f.read().splitlines())

# Load positive and negative words from MasterDictionary directory
master_dict_path = '/Users/tapastalati/Downloads/20211030 Test Assignment/MasterDictionary'
positive_words_path = os.path.join(master_dict_path, 'positive-words.txt')
negative_words_path = os.path.join(master_dict_path, 'negative-words.txt')

# Read positive and negative words
positive_words = set(open(positive_words_path, 'r', encoding='latin-1').read().splitlines())
negative_words = set(open(negative_words_path, 'r', encoding='latin-1').read().splitlines())


# Function to calculate number of syllables in a word
def syllable_count(word):
    # Implementation of syllable counting can vary, here's a basic approach
    count = 0
    vowels = 'aeiouy'
    word = word.lower()
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
    if word.endswith('e'):
        count -= 1
    if count == 0:
        count += 1
    return count

# Function to compute text analysis variables
def compute_text_analysis(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    
    # Remove punctuation
    words = [word for word in words if word not in string.punctuation]
    
    # Remove stop words
    words = [word for word in words if word.lower() not in stop_words]
    
    # Calculate word count
    word_count = len(words)
    
    # Calculate sentence count
    sentence_count = len(sent_tokenize(text))
    
    # Calculate average sentence length
    average_sentence_length = word_count / sentence_count
    
    # Calculate complex word count
    complex_word_count = sum(1 for word in words if syllable_count(word) > 2)
    
    # Calculate percentage of complex words
    percentage_complex_words = (complex_word_count / word_count) * 100
    
    # Calculate fog index
    fog_index_value = 0.4 * (average_sentence_length + percentage_complex_words)
    
    # Calculate average number of words per sentence
    avg_words_per_sentence = word_count / sentence_count
    
    # Calculate syllable per word
    syllable_per_word = sum(syllable_count(word) for word in words) / word_count
    
    # Calculate personal pronouns count
    personal_pronouns = sum(1 for word in words if word.lower() in ['i', 'me', 'my', 'mine', 'myself', 'we', 'us', 'our', 'ours', 'ourselves'])
    
    # Calculate average word length
    avg_word_length = sum(len(word) for word in words) / word_count
    
    # Perform sentiment analysis using TextBlob
    blob = TextBlob(text)
    positive_score = sum(1 for word in blob.words if word in positive_words)
    negative_score = sum(1 for word in blob.words if word in negative_words)
    polarity_score = blob.sentiment.polarity
    subjectivity_score = blob.sentiment.subjectivity
    
    # Return the computed variables
    return {
        'WORD COUNT': word_count,
        'AVG SENTENCE LENGTH': average_sentence_length,
        'COMPLEX WORD COUNT': complex_word_count,
        'PERCENTAGE OF COMPLEX WORDS': percentage_complex_words,
        'SYLLABLE PER WORD': syllable_per_word,
        'PERSONAL PRONOUNS': personal_pronouns,
        'AVG WORD LENGTH': avg_word_length,
        'FOG INDEX': fog_index_value,
        'POSITIVE SCORE': positive_score,
        'NEGATIVE SCORE': negative_score,
        'POLARITY SCORE': polarity_score,
        'SUBJECTIVITY SCORE': subjectivity_score,
        'AVG NUMBER OF WORDS PER SENTENCE': avg_words_per_sentence
    }
# Read the list of text files
text_files = [file for file in os.listdir() if file.endswith('.txt')]

# Create an empty DataFrame to store the results
results_df = pd.DataFrame(columns=['URL_ID','URL', 'WORD COUNT', 'AVG SENTENCE LENGTH', 'COMPLEX WORD COUNT',
                                   'PERCENTAGE OF COMPLEX WORDS', 'SYLLABLE PER WORD', 'PERSONAL PRONOUNS',
                                   'AVG WORD LENGTH', 'FOG INDEX', 'POSITIVE SCORE', 'NEGATIVE SCORE',
                                   'POLARITY SCORE', 'SUBJECTIVITY SCORE', 'AVG NUMBER OF WORDS PER SENTENCE'])
# Perform text analysis for each text file
for text_file in text_files:
    # Extract URL_ID from the file name
    url_id = os.path.splitext(os.path.basename(text_file))[0]
    
    # Fetch the corresponding URL from the dataframe based on URL_ID
    url = df.loc[df['URL_ID'] == url_id, 'URL'].values[0]
    
    with open(text_file, 'r') as file:
        text = file.read()
        analysis_result = compute_text_analysis(text)
        
        # Add the URL and URL_ID to the analysis result
        analysis_result['URL_ID'] = url_id
        analysis_result['URL'] = url
        
        # Append the analysis result to the results DataFrame
        results_df = pd.concat([results_df, pd.DataFrame(analysis_result, index=[0])], ignore_index=True)




# Reorder the DataFrame columns
results_df = results_df[['URL_ID', 'URL', 'POSITIVE SCORE', 'NEGATIVE SCORE', 'POLARITY SCORE', 'SUBJECTIVITY SCORE',
                         'AVG SENTENCE LENGTH', 'PERCENTAGE OF COMPLEX WORDS', 'FOG INDEX', 
                         'AVG NUMBER OF WORDS PER SENTENCE', 'COMPLEX WORD COUNT', 'WORD COUNT',
                         'SYLLABLE PER WORD', 'PERSONAL PRONOUNS', 'AVG WORD LENGTH']]

# Display or save the results
print(results_df)

# Write the results to an Excel file
results_df.to_excel('text_analysis_results.xlsx', index=False)
```
